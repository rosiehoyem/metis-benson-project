```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image

# enables inline plots, without it plots don't show up in the notebook
%matplotlib inline
```

```python
df = pd.read_csv('data/subway_tract_intersect_0.csv')
df1 = pd.read_csv('data/turnstile_160507 copy.txt')
```

```python
capitalizer = lambda x: x.upper()
df['name'] = df['name'].apply(capitalizer)

clean_ave = lambda x: re.sub('AVE','AV', x)
df['name'] = df['name'].apply(clean_ave)    
```

```python
# def clean_numbers(x):
#     matches = re.search(r'([\b\d+])([RD])', x)
#     for match in matches:
#         match.group(2) = ''

clean_numbers1 = lambda x: re.sub(r'(\b\d+)(RD\b)',r'\1', x)
df['name'] = df['name'].apply(clean_numbers1)

clean_numbers2 = lambda x: re.sub(r'(\b\d+)(TH\b)',r'\1', x)
df['name'] = df['name'].apply(clean_numbers2)

clean_numbers3 = lambda x: re.sub(r'(\b\d+)(ST\b)',r'\1', x)
df['name'] = df['name'].apply(clean_numbers3)

clean_numbers4 = lambda x: re.sub(r'(\b\d+)(ND\b)',r'\1', x)
df['name'] = df['name'].apply(clean_numbers4)

clean_dashes = lambda x: re.sub(r'(\w*)(\s-\s)(\w*)',r'\1-\3', x)
df['name'] = df['name'].apply(clean_dashes)

df.sample(40)
```

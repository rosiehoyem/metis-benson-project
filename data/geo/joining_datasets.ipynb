```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
stations = pd.read_csv('subway_station_tracts.csv')
stem_women = pd.read_csv('stem_women.csv')

df = pd.merge(stations, stem_women, on='BoroCT2010')
```

```python
df
```

```python
df.plot.bar(x='BoroCT2010', y='stem_women');
```

```python
df_sorted = df.sort_values('stem_women', ascending=False)
```

```python
df_sorted.plot.bar(x='BoroCT2010', y='stem_women');
```
